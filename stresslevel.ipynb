What are the key environmental factors affecting student wellbeing?


```python
# Radar chart for environmental factors
environmental_factors = ['noise_level', 'living_conditions', 'basic_needs']
high_stress = df[df['stress_level'] == 2][environmental_factors].mean()
low_stress = df[df['stress_level'] == 0][environmental_factors].mean()

angles = np.linspace(0, 2*np.pi, len(environmental_factors), endpoint=False)
angles = np.concatenate((angles, [angles[0]]))

high_stress_values = np.concatenate((high_stress.values, [high_stress.values[0]]))
low_stress_values = np.concatenate((low_stress.values, [low_stress.values[0]]))

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(projection='polar'))
ax.plot(angles, high_stress_values, 'o-', linewidth=2, label='High Stress', color='red')
ax.plot(angles, low_stress_values, 'o-', linewidth=2, label='Low Stress', color='green')
ax.fill(angles, high_stress_values, alpha=0.25, color='red')
ax.fill(angles, low_stress_values, alpha=0.25, color='green')
ax.set_xticks(angles[:-1])
ax.set_xticklabels([f.replace('_', ' ').title() for f in environmental_factors])
ax.legend()
plt.title('Environmental Factors: High vs Low Stress Students', y=1.08, fontsize=14, fontweight='bold')
plt.show()
```

Insight: Radar chart demonstrates high-stress students face worse environmental conditions across all factors, with living conditions showing the largest disparity.

Outlier Analysis And Handling

```python
numerical_cols = [col for col in df.columns if df[col].dtype in ['int64', 'float64']]
cols_per_row = 3
rows_needed = (len(numerical_cols) + cols_per_row - 1) // cols_per_row

# Subplot 
fig, axes = plt.subplots(nrows=rows_needed, ncols=cols_per_row, figsize=(15, rows_needed * 4))
axes = axes.flatten()

# boxplot for each column
for i, col in enumerate(numerical_cols):
    sns.boxplot(x=df[col], ax=axes[i])
    axes[i].set_title(f'Boxplot of {col}')
    axes[i].grid(False) 

# hiding extra axes if less charts present
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

Outlier Handling

```python
numerical_cols = [col for col in df.columns if df[col].dtype in ['int64', 'float64']]

# capping of each column
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    
    # Winsorization: clip values to lower/upper limit
    df[col] = np.where(df[col] < lower_limit, lower_limit, df[col])
    df[col] = np.where(df[col] > upper_limit, upper_limit, df[col])

print("Outliers handled with Winsorization using IQR:")
print(df.head()) 
```

Advanced Statistical Analysis with Visualizations

```python
# Risk factor identification and visualization
df['risk_category'] = df['stress_level'].map({0: 'Low Risk', 1: 'Medium Risk', 2: 'High Risk'})

# Calculate risk differences
numeric_cols = df.select_dtypes(include=[np.number]).columns.drop('stress_level')
risk_analysis = []

for col in numeric_cols:
    low_risk_mean = df[df['stress_level'] == 0][col].mean()
    high_risk_mean = df[df['stress_level'] == 2][col].mean()
    risk_difference = high_risk_mean - low_risk_mean
    risk_analysis.append({
        'Factor': col,
        'Low_Risk_Mean': low_risk_mean,
        'High_Risk_Mean': high_risk_mean,
        'Risk_Difference': risk_difference
    })

risk_df = pd.DataFrame(risk_analysis).sort_values('Risk_Difference', key=abs, ascending=False)

# Visualization of risk factors
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Top risk differentiators
top_risks = risk_df.head(10)
sns.barplot(data=top_risks, x='Risk_Difference', y='Factor', ax=axes[0,0], 
            palette=['red' if x > 0 else 'green' for x in top_risks['Risk_Difference']])
axes[0,0].set_title('Top 10 Risk Factor Differences (High Risk - Low Risk)')
axes[0,0].axvline(x=0, color='black', linestyle='-', alpha=0.3)

# Risk category distribution
risk_counts = df['risk_category'].value_counts()
axes[0,1].pie(risk_counts.values, labels=risk_counts.index, autopct='%1.1f%%', 
              colors=['green', 'orange', 'red'], startangle=90)
axes[0,1].set_title('Risk Category Distribution')

# Stacked bar chart for multiple factors
key_factors = ['anxiety_level', 'depression', 'sleep_quality', 'social_support', 'academic_performance']
risk_comparison = df.groupby('risk_category')[key_factors].mean()
risk_comparison.plot(kind='bar', ax=axes[1,0], colormap='viridis', width=0.8)
axes[1,0].set_title('Average Scores by Risk Category')
axes[1,0].legend(bbox_to_anchor=(1.05, 1), loc='upper left')
axes[1,0].set_xticklabels(axes[1,0].get_xticklabels(), rotation=45)

# Radar chart for risk profiles
categories = ['anxiety_level', 'depression', 'sleep_quality', 'social_support']
angles = np.linspace(0, 2*np.pi, len(categories), endpoint=False)
angles = np.concatenate((angles, [angles[0]]))

ax_radar = plt.subplot(2, 2, 4, projection='polar')
for risk_level, color in zip(['Low Risk', 'Medium Risk', 'High Risk'], ['green', 'orange', 'red']):
    values = df[df['risk_category'] == risk_level][categories].mean().values
    values = np.concatenate((values, [values[0]]))
    ax_radar.plot(angles, values, 'o-', linewidth=2, label=risk_level, color=color)
    ax_radar.fill(angles, values, alpha=0.25, color=color)

ax_radar.set_xticks(angles[:-1])
ax_radar.set_xticklabels([cat.replace('_', ' ').title() for cat in categories])
ax_radar.set_title('Risk Profile Comparison (Radar Chart)')
ax_radar.legend()

plt.tight_layout()
plt.show()
```
